```python
from pychunk.chunkers.python_chunker import PythonChunker

files_path = ["pychunk/"]
chunker = PythonChunker(files_path=files_path)
```

```python
nodes = chunker.find_relationships()
```

```python
from pychunk.nodes.types import NodeRelationshipType

all_nodes = {}
for file_path, node_of_file in nodes.items():
    for id, node in node_of_file.items():
        all_nodes[id] = node

for _, nodes_file in nodes.items():
    for id, node in nodes_file.items():
        if len(node.filter_relationships(NodeRelationshipType.OTHER)) > 5:
            print(node)
            print("Text of this node: \n", node.content)  
            node_relationships = node.filter_relationships(NodeRelationshipType.OTHER)
            print("Relationships of this node: \n")
            for n, rel in enumerate(node_relationships):
                print(f"Relationship {n+1} -> {rel}: \n", all_nodes[rel].content)
                print("\n")
            print("\n")
            print("-"*100)
            print(" ")
    
```

```python
for node_id, node in all_nodes.items():
    print(f"Node id: ", node_id + " from ", node.file)
    print(f"Content of node type: {str(node.node_type.value)}: \n")
    print(node.content)
    print("-"*100)
```
